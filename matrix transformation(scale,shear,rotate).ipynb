```python
import numpy as np
from scipy import ndimage
from matplotlib import pyplot as plt
%matplotlib inline
from PIL import Image
a=Image.open('mohanlal.jpg')
b=np.array(a)

x=ndimage.rotate(b,45,cval=225)
d=Image.fromarray(x)
plt.imshow(d)
# a.show()
# p=a.rotate(45)
# p.show()
```

```python
c=ndimage.zoom(b,(.25,.25,1))
e=Image.fromarray(c)
plt.imshow(e)
```

```python
h,w,c=b.shape
k=np.array([[1,.5,0],[0,1,0],[0,0,1]])
l=ndimage.affine_transform(b,k)
r=Image.fromarray(l)
plt.imshow(r)
```
