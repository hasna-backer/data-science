```python
#dataset cleaning 
import pandas as pd
a=pd.read_csv('vt.csv')
#to see as full content
print(a.to_string())
a.duplicated().any()
m=a['Calories'].mean()
a['Calories'].fillna(m,inplace=True)
print('\n')
print(a.to_string())
a.loc[3,'Duration']=45
print('\n')
print(a.to_string())
a['Date']=pd.to_datetime(a['Date'])
print('\n')
print(a.to_string())
a['Pulse']=a['Pulse'].astype('float')
print('\n')
print(a.to_string())
print('\n')
a.duplicated()
a.drop_duplicates(inplace=True)
print('\n')
print(a.to_string())
```
