```python
while True:
    
    print("\nOPERATIONS\n1.add\n2.subtract\n3.multiplication\n4.division\n5.exit")
    ch=int(input("ente your choice:\n\n"))
    if ch==1:
        a=int(input("enter first number:"))
        b=int(input("enter second number:"))
        c=a+b
        print("\n",a,"+",b,"=",c)
    elif ch==2:
        a=int(input("enter first number:"))
        b=int(input("enter second number:"))
        c=a-b
        print("\n",a,"-",b,"=",c)
    elif ch==3:
        a=int(input("enter first number"))
        b=int(input("enter second number"))
        c=a*b
        print("\n",a,"*",b,"=",c)
    elif ch==4:
        a=int(input("enter first number:"))
        b=int(input("enter second number:"))
        if b==0:
            print("division is not possible")
        else:
            c=a/b
            print("\n",a,"/",b,"=",c)
    elif ch==5:
        break
        
    else:
        print("wrong input")
        
    
```
